```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Solar" : [1200, 1500, 1300],
    "Wind " : [3400, 3600,3200],
    "Hydropower" : [2900, 3100, 2800],
    "Biomass" : [2500, 2700, 2400]
}

df = pd. DataFrame(data)

sns.pairplot(df)
plt.show()
```

```python
correleation_matrix = df.corr()

sns.heatmap(correleation_matrix, annot = True, cmap= 'coolwarm', linewidths = 0.5)

plt.title("Correleation Between Energy and Sources")
plt.show()
```

```python
sns.boxplot(data = df)

plt.title("Distribution of Energy Consumption by Source")
plt.show()
```

```python
sns.violinplot(data = df)

plt.title("Violin Plot of Energy Consumption Distributions")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

energy_values = [100, 200, 300, 400, 500]
carbon_emission = [10, 20, 25, 40, 50]

df_reg = pd.DataFrame(
    {'Energy Consumption' : energy_values,
     'Carbon Emissions' : carbon_emission
    }
                     )
sns.regplot(x = 'Energy Consumption', y= 'Carbon Emissions', data = df_reg)
plt.title('Relationship between Energy Consumption and Carbon Emissions')
plt.show()
```

```python
import numpy as np
from mpl_toolkits import mplot3d

fig = plt.figure()

ax = plt.axes(projection = '3d')

z = np.linspace(0, 1, 1000)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x, y, z, 'green')
ax.set_title('3D line plot')
plt.show()
```
